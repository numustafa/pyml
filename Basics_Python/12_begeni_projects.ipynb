### Guess the no by Computer

```python
# Guess the no
import random

def guess(y):
    rand_no = random.randint(1,y)
    guess = 0
    while guess != rand_no:
        guess = int(input(f"Guess the no b/w 1 & {y}"))
        if guess < rand_no:
            print(f"Sorry! {guess} is too low")
        elif guess > rand_no:
            print(f"Sorry! {guess} is too high")
    print(f"Yay! {rand_no} is a correct guess")
```

```python
guess(120)
```

### Rock, Paper Scisors

```python
# Rock, paper, Scisors (Rules for win: p>r, s>p, r>s)
import random

def play():
    usr = input("Your Choice?: 'r' for rock, 'p' for paper & 's' for scisors")
    comp = random.choice(["r",'p','s'])

    if usr == comp:
        return "tie"
    
    if is_win(usr, comp):
        return "Win"
    return "You Lost!!"

def is_win(player, oponnent):
    if (player == 'r' and oponnent == 's') or (player == 's' and oponnent == 'p') or (player == 'p' and oponnent == 'r'):
        return True
```

```python
play()
```

### Binary Search

```python
# Binary Search (Divide and conquer algo helps to search ordered list faster)
# if yes: return the index, else return -1
def binary_search(l, target, low =None, high =None):
    if low is None:
        low = 0
    if high is None:
        high = len(l)-1
    if high<low:
        return -1
    mid_pt = (low + high)//2

    if l[mid_pt] == target:
        return mid_pt
    elif target < l[mid_pt]:
         return binary_search(l, target, high = mid_pt-1)
    else:
        return binary_search(l, target, low = mid_pt+1)

  
```

```python
l = [1,3,5,10,12]     # this is a sorted list
target = 10
binary_search(l,target)  # 3rd index
```
