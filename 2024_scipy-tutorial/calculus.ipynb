# Calculus
## 1- Differentiation
Different than Numpy

```python
import numpy as np
import scipy
import matplotlib.pyplot as plt
from scipy.misc import derivative
```

```python
# example
x = np.linspace(0,1,1000)
def f(x):
    return (x**2)*np.sin(2*x)*np.exp(-x)

```

```python
plt.plot(x,f(x))
```

```python
# derivative of f(x)
dfdx = derivative(f, x, dx = 1e-6)       # with small dx

plt.plot(x, f(x))
plt.plot(x, dfdx)
```

```python
# higher order derivative
df_sq_dx = derivative(f, x, dx = 1e-6, n = 2)
plt.plot(x, f(x))
plt.plot(x, dfdx)
plt.plot(x, df_sq_dx)
```

## 2- Integration
Unlike Numpy, it dont use cumsum, instead has a function for it.

```python
from scipy.integrate import quad
```

```python
def integrand(x):
    return x**2*np.sin(2*x)*np.exp(-x)

integral, integral_error = quad(integrand, 0,1)
```

```python
integral
```

```python
integral_error
```

```python
from scipy.integrate import dblquad
```

```python
def integrand(x,y):
    return np.sin(x+y**2)

def lwr_y(x):
    return -x
def upr_y(x):
    return x**2

integral, integral_error = dblquad(integrand, 0,1, lwr_y, upr_y)
```

```python
integral
```

```python
integral_error
```
