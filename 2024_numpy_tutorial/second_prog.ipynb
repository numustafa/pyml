```python
import numpy as np
import matplotlib.pyplot as plt
```

## Examples

### Q-1


```python
# make a contour plot
N = 10000
x = np.linspace(-2,2,N)
y = np.linspace(-2,2,N)

```

```python
xv, yv = np.meshgrid(x,y)
f_xy = np.exp(-(xv**2 + yv**2)) * np.sin(xv)
```

```python

plt.contourf(xv,yv,f_xy, levels = 30)
plt.colorbar()
```

```python
# find the absolute value (vol = cumsum of f values * dx * dy) - earlier we used dx = x[1] - x[0]
absol = np.abs(f_xy.ravel()) 
```

```python
vol = np.sum(absol) * np.diff(x)[0] * np.diff(y)[0]
vol
```

```python
# find the vol in the region sqrt(x**2 + y**2)>0.5
# find proper indexation
mask = (xv**2 + yv**2) > 0.5**2

f_xy_masked = f_xy[mask]          # Apply the mask to the function values
absol_masked = np.abs(f_xy_masked)
vol_masked = np.sum(absol_masked) * np.diff(x)[0] * np.diff(y)[0]
print("Volume in region where √(x² + y²) > 0.5:", vol_masked)
```
