### integral & derivative

```python
x = np.linspace(1,10,100)
def f(x):
    return (1/x**2) * np.sin(x)

y = f(x)
plt.plot(x,y)
```

```python
# Derivative
dydx = np.gradient(y, x)

plt.plot(x,y)
plt.plot(x, dydx)
```

## Example


```python
N = 10000
x= np.linspace(0,10, N+1)
def f(x):
    return np.exp(-x/10)*np.sin(x)

y = f(x)

# 1. Plot the fn y vs x
plt.plot(x,y)
```

```python
# Compute the mean & std of y, for x values in [4,7]
new_1 = x[x<=7]
new_x = new_1[new_1>=4]
new_x
```

```python
y_new = f(new_x)         # or we can simpley use y[(x>=4)*(x<=7)]
mean = np.mean(y_new)
std = np.std(y_new)
print("Mean & std of y in x = [4,7] is: ", mean, std)
```

```python
# 80% of y-values are less than a marked no in new_x
np.percentile(y_new, 80)
```

```python
# plot dy/dx vs x
dydx = np.gradient(y,x)
plt.plot(x,y)
plt.plot(x,dydx)
```

```python
# Find the location where dydx == 0
dydx[dydx == 0]      # no place where dydx = 0
```

```python
# focus on where the derivative switches place, and get those particular x-values

"""
1- analyse the dydx array, starts with index 0 - 10000 (incl)
2- we want those index where dydx array changes signs
3- Trick: 
    take dydx[1:] - this eliminates the 0th index
    take dydx[:-1] - this eliminales the 10000th index
    Multiply both the 10000 values (+*+ = + & -*- = +), take indices where values are negative and 
    use tose indices to get values from x-array (to match the indices, we take x[1:])
"""
x[1:][dydx[1:] * dydx[:-1] < 0]

```

## Q-3


```python
N = 10000
theta = np.linspace(0, 2*np.pi, N)
r_theta = 1+ (3/4) * np.sin(3*theta)

# convert polar coordinates into cartisian coordinates
x = r_theta * np.cos(theta)
y = r_theta * np.sin(theta)
```

```python
# Make a plot of the flower
plt.plot(x, y)
```

```python
# Compute the area
A = 1/2 * (np.sum(r_theta**2)) * (theta[1] - theta[0])
A
```

```python
# Compute the arc-length
drd_theta = np.gradient(r_theta, theta)
d_theta = theta[1] - theta[0]
L = np.sum(np.sqrt((r_theta**2) + (drd_theta)**2))*d_theta
L
```

## System of Eqn


```python
# setup a Matrix
A = np.array([[3,2,1],[5,-5,4],[6,0,1]])
b = np.array([4,3,0])
```

```python
# Solve the system
np.linalg.solve(A,b)
```
